```python
import time

def safe_invoke(agent, payload, config=None, retries=6):
    delay = 2
    last_err = None
    for _ in range(retries):
        try:
            return agent.invoke(payload, config=config)
        except Exception as e:
            msg = str(e)
            last_err = e
            if "RESOURCE_EXHAUSTED" in msg or "429" in msg or "quota" in msg.lower():
                time.sleep(delay)
```

```python
import os
import datetime
from dataclasses import dataclass
from typing import Dict, Any

# -------------------------
# TOOLS (mock tools)
# -------------------------
def get_weather(city: str) -> str:
    """Return mock weather information for a given city."""
    return f"Mock weather for {city}: Sunny."

def search_internet(query: str) -> str:
    """Perform a mock internet search and return sample results."""
    results = [
        {
            "title": "LangGraph Memory",
            "url": "https://langchain-ai.github.io/langgraph/",
            "snippet": "Short-term memory using InMemorySaver/checkpointers.",
        },
        {
            "title": "LangChain Agents",
            "url": "https://python.langchain.com/",
            "snippet": "Agents, tools, and prompting patterns.",
        },
    ]
    out = [f"Search Results for: {query}\n"]
    for i, r in enumerate(results, 1):
        out.append(f"{i}. {r['title']} - {r['url']}\n   {r['snippet']}\n")
    return "\n".join(out)

def send_email(to: str, subject: str, body: str) -> str:
    """Send a mock email (simulation only, no real email is sent)."""
    t = datetime.datetime.now().isoformat(timespec="seconds")
    return (
        "EMAIL SENT (MOCK)\n"
        f"Time: {t}\n"
        f"To: {to}\n"
        f"Subject: {subject}\n"
        f"Body:\n{body}\n"
    )

# -------------------------
# SHORT-TERM MEMORY (per thread_id)
# -------------------------
MEMORY: Dict[str, Dict[str, Any]] = {}

def mem_get(thread_id: str, key: str, default=None):
    return MEMORY.get(thread_id, {}).get(key, default)

def mem_set(thread_id: str, key: str, value: Any):
    if thread_id not in MEMORY:
        MEMORY[thread_id] = {}
    MEMORY[thread_id][key] = value

# -------------------------
# OFFLINE AGENT (NO API)
# - Creates "responses" + uses tools
# -------------------------
@dataclass
class OfflineAgent:
    thread_id: str

    def invoke(self, user_message: str) -> str:
        msg = user_message.lower().strip()

        # Save name
        if "my name is" in msg:
            name = user_message.split("is", 1)[-1].strip().strip(".")
            mem_set(self.thread_id, "name", name)
            return f"Got it. I will remember your name as {name}."

        # Ask name
        if "what is my name" in msg or "do you know my name" in msg:
            name = mem_get(self.thread_id, "name", None)
            return (
                f"Yes, your name is {name}."
                if name
                else "I don't know your name yet. Please tell me: 'My name is ...'"
            )

        # Weather
        if "weather" in msg:
            city = "San Francisco"
            if "in " in msg:
                city = user_message.split("in", 1)[-1].strip().strip("?").strip(".")
            tool_out = get_weather(city)
            return f"I used the weather tool.\n{tool_out}"

        # Internet Search (FIXED INDENTATION ✅)
        if "search" in msg or "internet" in msg:
            q = user_message.strip()

            # remove common command phrases
            for phrase in [
                "Search the internet for", "search the internet for",
                "Search internet for", "search internet for",
                "Search the internet", "search the internet",
                "Search internet", "search internet",
                "Search for", "search for",
                "Search", "search", "internet",
            ]:
                q = q.replace(phrase, "")

            query = q.strip().strip(":").strip("-").strip()
            if not query:
                query = "LangGraph short term memory"

            tool_out = search_internet(query)
            return f"I used the internet search tool.\n{tool_out}"

        # Email
        if "send email" in msg or "send an email" in msg:
            to = "test@example.com"
            subject = "Demo"
            body = "Hello"
            tool_out = send_email(to, subject, body)
            return f"I used the email tool.\n{tool_out}"

        return "I can help with: weather, internet search, send email, and remembering your name."

# -------------------------
# REAL GEMINI AGENT (OPTIONAL)
# Needs: pip install langchain langchain-google-genai langgraph
# and a valid GOOGLE_API_KEY
# -------------------------
def try_real_agent(thread_id: str):
    key = os.environ.get("GOOGLE_API_KEY")
    if not key:
        return None

    try:
        from langchain_google_genai import ChatGoogleGenerativeAI
        from langchain.agents import create_agent
        from langgraph.checkpoint.memory import InMemorySaver
        from langchain_core.tools import tool

        @tool
        def t_get_weather(city: str) -> str:
            """Return mock weather information for a given city."""
            return get_weather(city)

        @tool
        def t_search_internet(query: str) -> str:
            """Perform a mock internet search and return sample results."""
            return search_internet(query)

        @tool
        def t_send_email(to: str, subject: str, body: str) -> str:
            """Send a mock email (simulation only)."""
            return send_email(to, subject, body)

        @tool
        def set_my_name(name: str) -> str:
            """Save user's name in short-term memory for this thread."""
            mem_set(thread_id, "name", name)
            return f"Saved name: {name}"

        @tool
        def get_my_name() -> str:
            """Get user's saved name from short-term memory for this thread."""
            return mem_get(thread_id, "name", "Unknown")

        llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)
        checkpointer = InMemorySaver()

        agent = create_agent(
            model=llm,
            tools=[t_get_weather, t_search_internet, t_send_email, set_my_name, get_my_name],
            system_prompt="You are a helpful assistant. Use tools when needed.",
            checkpointer=checkpointer,
        )
        return agent
    except Exception:
        return None

# -------------------------
# DEMO RUN (for teacher)
# -------------------------
if __name__ == "__main__":
    THREAD_ID = "Chinto123"

    offline = OfflineAgent(thread_id=THREAD_ID)

    print("\n=== OFFLINE AGENT DEMO (NO API) ===")
    print(offline.invoke("My name is chinto"))
    print(offline.invoke("What is my name?"))
    print(offline.invoke("What is the weather in London?"))
    print(offline.invoke("Search the internet for LangGraph short term memory"))
    print(offline.invoke("Send email"))

    real = try_real_agent(thread_id=THREAD_ID)
    if real:
        print("\n=== REAL GEMINI AGENT DEMO (API REQUIRED) ===")
        r = real.invoke(
            {"messages": [("user", "My name is chinto. Remember it.")]},
            config={"configurable": {"thread_id": THREAD_ID}},
        )
        print(r["messages"][-1].content)

        r = real.invoke(
            {"messages": [("user", "What is my name?")]},
            config={"configurable": {"thread_id": THREAD_ID}},
        )
        print(r["messages"][-1].content)
    else:
        print("\n(Real Gemini agent not started: missing/invalid API key or packages.)")
```
