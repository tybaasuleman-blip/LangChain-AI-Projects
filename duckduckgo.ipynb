```python
from ddgs import DDGS

def search_internet(query, max_results=5):
    """
    Custom function to search DuckDuckGo and return a clean string of results.
    """
    results_text = ""
    
    with DDGS() as ddgs:
        # text() function parameters:
        # keywords: the search string
        # region: 'wt-wt' (worldwide) or 'us-en'
        # max_results: limits the number of items returned
        search_gen = ddgs.text(query, region="wt-wt", max_results=max_results)
        
        for i, r in enumerate(search_gen, 1):
            title = r.get('title', 'No Title')
            link = r.get('href', 'No Link')
            body = r.get('body', 'No Description')
            
            # Format each result into a readable block
            results_text += f"[{i}] {title}\nURL: {link}\nSummary: {body}\n\n"
            
    return results_text

# --- Example Usage ---
my_query = "What is the newest feature in Python 3.14?"
search_data = search_internet(my_query, max_results=3)

print("--- SEARCH OUTPUT ---")
print(search_data)
```

```python
import serpapi

# Configuration
SERP_API_KEY = "d4c3bc5e4d36d82f8847e9bfb259b897d255cdb277a60988f6ec91600549630f"

class MemoryAgent:
    def __init__(self):
        # 1. THE IN-MEMORY STORE (RAM)
        self.memory = {
            "history": [],      # List of past user/assistant turns
            "context": {},      # Facts like user name, current topic
            "max_turns": 10     # Limits memory so it doesn't get too big
        }
        self.client = serpapi.Client(api_key=SERP_API_KEY)

    def add_to_memory(self, role, content):
        self.memory["history"].append({"role": role, "content": content})
        # Keep memory clean (Last 10 messages)
        if len(self.memory["history"]) > self.memory["max_turns"]:
            self.memory["history"].pop(0)

    def search_tool(self, query):
        """A simple search tool the agent can use."""
        print(f"🤖 Agent is searching: {query}...")
        results = self.client.search({"q": query, "engine": "google"})
        if "organic_results" in results:
            return results["organic_results"][0].get("snippet", "No info found.")
        return "Search failed."

    def run(self, user_input):
        # 2. LOGIC: Update memory with the new question
        self.add_to_memory("user", user_input)
        
        # 3. ACTION: Perform a task (e.g., search)
        # In a real agent, the LLM would decide to call this.
        info = self.search_tool(user_input)
        
        # 4. RESPONSE: Combine info and memory
        response = f"Based on my memory of your question '{user_input}', I found: {info}"
        self.add_to_memory("assistant", response)
        
        return response

# --- Test the Agent ---
my_agent = MemoryAgent()
print(my_agent.run("Who is Zeeshan Ali P2P Clouds?"))
print(f"\nMemory Content: {len(my_agent.memory['history'])} messages stored.")
```
